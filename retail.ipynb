```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('OnlineRetail.csv', encoding='unicode_escape')
df.head()
```

```python
# 1. Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
df.shape
df.info()
```

```python
# 2. Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; nhận
# giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;
# # nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;
df['Date'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M').dt.day
df['Month'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M').dt.month
df['Year'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M').dt.year
df.head()
```

```python
condition = [df['Month'].isin([1,2,3]), df['Month'].isin([4,5,6]),\
     df['Month'].isin([7,8,9]), df['Month'].isin([10,11,12])]
choicelist = [1,2,3,4]
df['Previous'] = np.select(condition,choicelist)
df.head()
```

```python
# Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice
df['Amount'] = df['Quantity']*df['UnitPrice']
df.head()
```

```python
# 4. Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.
condition = [(df['Country'] == 'United Kingdom') & (df['Previous'] == 4),df['Country'] == 'France']
choicelist = [10,5]
df['Discount'] = np.select(condition,choicelist)
df.head()
```

```python
# 5. Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.
df['Total'] = df['Amount']*(1 - df['Discount']/100)
df.head()
```
